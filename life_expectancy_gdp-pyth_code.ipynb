```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
readed_data= pd.read_csv("all_data.csv")
```

```python
print(readed_data.head())
```

```python
readed_data["Count_Value"]=readed_data.Year
readed_data.columns=["Country","Year","LEB_Years","GDP","Count_Value"]

groupte=readed_data.groupby("Country").Year.count().reset_index()
```

```python
unique_countries=groupte.Country
print(unique_countries)

for countq in unique_countries:
    plt.plot(readed_data.Year[readed_data.Country==countq],readed_data.GDP[readed_data.Country==countq])

plt.legend(unique_countries)
plt.xlabel("Year")
plt.ylabel("GDP")    
plt.title("Countries Yearly GDP")
plt.show()
plt.clf()

for countq in unique_countries:
    plt.plot(readed_data.Year[readed_data.Country==countq],readed_data.LEB_Years[readed_data.Country==countq])

plt.legend(unique_countries)
plt.ylabel("Life expectancy at birth (years)")
plt.xlabel("Year")    
plt.title("Countries Life expectancy at birth Compare")
plt.show()
plt.clf()
```

```python
zimbabwe_datas=readed_data[readed_data.Country=="Zimbabwe"].reset_index()
print(zimbabwe_datas.head())
```

```python
'''plt.plot(zimbabwe_datas.Year,zimbabwe_datas.GDP,color="purple",label="GDP")
plt.plot(zimbabwe_datas.Year,zimbabwe_datas.LEB_Years,color="yellow",label="LEB")
plt.legend()
'''
sns.scatterplot(x="GDP",y="LEB_Years",hue="Year",data=zimbabwe_datas)
plt.title("Zimbabwe DGP-Year Compare Figure")
plt.show()
```

```python
#general county check with heatmap for GDP
generic_country_year = readed_data.reset_index().pivot(columns="Year", index="Country", values="GDP")
plt.figure(figsize=(14,12))
plt.subplot(2,1,1)

sns.heatmap(generic_country_year, cbar_kws={"label": "Total GDP"})
plt.title("Countries GDP's Compared with Years")
plt.ylabel("Country")
plt.xlabel("Year")

#general county check with heatmap for LEB
plt.subplot(2,1,2)
generic_country_year = readed_data.reset_index().pivot(columns="Year", index="Country", values="LEB_Years")

sns.heatmap(generic_country_year, cbar_kws={"label": "Total Life expectancy at birth (years)"})
plt.title("Countries Life expectancy at birth (years)'s Compared with Years")
plt.ylabel("Country")
plt.xlabel("Year")
plt.savefig("Total_Compare.png")
plt.show()
```
